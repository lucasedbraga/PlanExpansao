### 5) a) Com base na conta de energia elétrica é possível realizar alguma análise com relação ao fator de potência da instalação tanto Fora Ponta e quanto Ponta? Justifique suas afirmações. b) A medição do fator de potência é horária mas, seria possível calcular um fator de potência médio (HFP e FP) com os dados fornecidos na fatura de energia elétrica?

O fator de potência está baixo porque foi cobrado Energia Reativa (Valores Faturados). Outro local que podemos visualizar é no Registro de Energia em houve medição da Energia Reativa em que podemos calcular o fator de potencia pelo triangulo de potencias:



```python
import math

def calc_fp(P,Q,S):
    fp = {}
    fp['cosfi'] = P/S
    #fp['angfi'] = math.degrees(math.acos(P/S))
    return fp    

P_fp = 255000
Q_fp = 6600
S_fp = 137400
P_p = 25200
Q_p = 1200
S_p = 12600

FP_fp = calc_fp(P_fp,Q_fp,S_fp)
print(f"Fator de Potência Fora Ponta = {FP_fp['cosfi']} \n"
     )#f"Angulo do Fator de Potência = {FP_fp['angfi']}", end='\n')

FP_p = calc_fp(P_p,Q_p,S_p)
print(f"Fator de Potência Ponta = {FP_p['cosfi']} \n"
     )#f"Angulo do Fator de Potência = {FP_p['angfi']}", end='\n')

```

### 6) Considerando apenas o mês atual de faturamento, efetue uma comparação entre as possíveis opções de enquadramento tarifário e determine qual a melhor opção para este consumidor. Determine as diferenças em termos absolutos (R$) e percentuais entre todas as modalidades tarifárias disponíveis no mercado cativo? Nesta análise, considere os valores de demanda contratada e ignore os custos com energia e demanda reativas, caso existirem. Utilize as tarifas disponíveis no arquivo xls anexo para o mês de Abril/2022.

```python
def calculo(valor, tarifa):
  return valor*tarifa

demanda_ponta = 714
demanda_fp = 774
energia_consumida_ponta = 25200
energia_consumida_fp = 255000

tarifa_demanda_ponta = 44.90
tarifa_demanda_fp = 14.86
tarifa_consumo_ponta = 637.66/1000
tarifa_consumo_fp = 488.64/1000
tarifa_unica = 14.86
tarifa_verde_consumo_p = 1725.06/1000
tarifa_verde_consumo_fp = 488.64/1000

demanda_contratada_p = 700
demanda_contratada_fp = 800

print(f"{'-'*40} \n THS AZUL A4 \n", end='\n')
tarifa_azul = []
tarifa_azul.append(calculo(demanda_ponta, tarifa_demanda_ponta))
tarifa_azul.append(calculo(demanda_fp, tarifa_demanda_fp))  
tarifa_azul.append(calculo(energia_consumida_ponta, tarifa_consumo_ponta))
tarifa_azul.append(calculo(energia_consumida_fp, tarifa_demanda_fp))
print(f'Tarifas = {tarifa_azul}',end='\n')
print(f'Valor da Fatura = {sum(tarifa_azul)}', end='\n')

```

### 7) Considerando os resultados da questão anterior, calcule o preço médio da energia (R$/kWh) para cada opção tarifária disponível no mercado cativo a este consumidor.

```python
print(f"{'-'*40} \n THS VERDE A4 \n", end='\n')
tarifa_verde = []
def max_consumo(medido, contratado):
  return max([medido, contratado])

tarifa_verde.append(calculo(max_consumo(demanda_ponta,demanda_contratada_p), tarifa_unica))
tarifa_verde.append(calculo(max_consumo(demanda_fp,demanda_contratada_fp), tarifa_unica))
tarifa_verde.append(calculo(energia_consumida_ponta, tarifa_verde_consumo_p))
tarifa_verde.append(calculo(energia_consumida_fp, tarifa_verde_consumo_fp))
print(f'Tarifas = {tarifa_verde}',end='\n')
print(f'Valor da Fatura = {sum(tarifa_verde)}', end='\n')
```
